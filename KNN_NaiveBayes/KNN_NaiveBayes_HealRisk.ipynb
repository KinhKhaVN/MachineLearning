```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

import time
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

fin = "Maternal Health Risk Data Set.csv"
df = pd.read_csv(fin)

df.head()
# frequency = df.groupby('RiskLevel').size()
# frequency
```

```python
columns = df.columns.tolist()
fig, axs = plt.subplots(2, 4, figsize = (12, 5))
axs = axs.flatten()


for i in range(len(columns)):
    sn.set(style = "darkgrid")
    
    axs[i].hist(df[columns[i]], bins = 10, color = "skyblue", edgecolor = "black")

    axs[i].set_title(f'Histogram for {columns[i]}', color = "red")

    axs[i].set_xlabel('Value')
    axs[i].set_ylabel('Frequency', color = "green")


plt.tight_layout()
plt.show()        


```
